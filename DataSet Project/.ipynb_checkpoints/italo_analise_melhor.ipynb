```python
# prompt: please make net_transfer_record a int for this um read the number and the 'm' can be read as million( example -16m would be -16.000.000)

def convert_transfer_to_int(value):
  if isinstance(value, str):
    if 'm' in value:
      return int(float(value.replace('m', '')) * 1000000)
  else:
    return value

df['net_transfer_record'] = df['net_transfer_record'].apply(convert_transfer_to_int)
df.head()
```

```python
# prompt: usando a tabela players faça um elbow method e depois um calinsk score com um sillhuette score

import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
from sklearn.metrics import calinski_harabasz_score

# Carrega os dados
df = pd.read_parquet("DataSet Project/transfermarkrt-dados-clean/players.parquet")

# Seleciona apenas as colunas numéricas para o clustering
numeric_columns = df.select_dtypes(include=['number']).columns
df_numeric = df[numeric_columns]

# Remove as colunas que não serão usadas no clustering
df_numeric = df_numeric.drop(['player_id', 'current_club_id'], axis=1)

# Preenche os valores nulos com a média da coluna
df_numeric = df_numeric.fillna(df_numeric.mean())

# Normaliza os dados
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

# Elbow Method
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(df_scaled)
visualizer.show()

# Calinski Harabasz Score
k_values = range(2, 10)
ch_scores = []
for k in k_values:
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(df_scaled)
  ch_scores.append(calinski_harabasz_score(df_scaled, kmeans.labels_))

# Encontra o valor de k com o maior Calinski Harabasz Score
best_k_ch = k_values[np.argmax(ch_scores)]
print("Melhor valor de k (Calinski Harabasz):", best_k_ch)

# Silhouette Score
fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4, 5]:
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  q, mod = divmod(i, 2)
  visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick', ax=ax[q-1][mod])
  visualizer.fit(df_scaled)
```
